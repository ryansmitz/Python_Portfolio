```python
#1 import dataframe
df_ords_prods_merge_2 = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'df_ords_prods_merge_2.pkl'))
```

```python
#find the aggregated mean for entire dataframe grouped by department id
df_ords_prods_merge_2.groupby('department_id').agg({
    'order_number': 'mean',
    'order_day_of_week': 'mean',
    'order_hour_of_day': 'mean',
    'days_since_prior_order': 'mean',
    'add_to_cart_order': 'mean',
    'reordered': 'mean',
    'prices': 'mean'
})
```

#3 Each figure in the table represents different aspects of the data. The order_number column allows us to compare order numbers between departments. The price column gives an idea of the average price of the purchased products in each department. This can provide information about price differences between departments or possibly buying habits. The hour_of_day column shows the average time of day that items are purchased, again by department. The hour_of_day column shows less variation than say, the price column. All departments show that items are purchased on average sometime between 1pm and 2pm.

```python
#4 create a max order column
df_ords_prods_merge_2['max_order'] = df_ords_prods_merge_2.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#eliminate max rows view
pd.options.display.max_rows = None
```

```python
#check the dataframe visually
df_ords_prods_merge_2.head(100)
```

```python
#4 create a loyalty flag column using the max order column
df_ords_prods_merge_2.loc[df_ords_prods_merge_2['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge_2.loc[(df_ords_prods_merge_2['max_order'] <= 40) & (df_ords_prods_merge_2['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge_2.loc[df_ords_prods_merge_2['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#visual check that the new column was created succesfully
df_ords_prods_merge_2.head(30)
```

```python
#check the value counts of the new loyalty flag
df_ords_prods_merge_2['loyalty_flag'].value_counts(dropna = False)
```

```python
#5 check the basic statistics of the product prices for each loyalty category
df_ords_prods_merge_2.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
#6 create a spending flag for each user based on the average price across all their orders
df_ords_prods_merge_2['average_purchase'] = df_ords_prods_merge_2.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# check the average_purchase column visually
df_ords_prods_merge_2.head(25)
```

```python
#create a purchase flag column with two categories: high spender and low spender
df_ords_prods_merge_2.loc[df_ords_prods_merge_2['average_purchase'] < 10, 'purchase_flag'] = 'Low Spender'
```

```python
df_ords_prods_merge_2.loc[df_ords_prods_merge_2['average_purchase'] >= 10, 'purchase_flag'] = 'High Spender'
```

```python
#visually check that the new column was created
df_ords_prods_merge_2.head(10)
```

```python
#7 create a column that shows a user's median order frequency
df_ords_prods_merge_2['median_order_frequency'] = df_ords_prods_merge_2.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#visually check the new column
df_ords_prods_merge_2.head(25)
```

```python
#7 create a median order frequency flag column with three categories
df_ords_prods_merge_2.loc[df_ords_prods_merge_2['median_order_frequency'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
df_ords_prods_merge_2.loc[(df_ords_prods_merge_2['median_order_frequency'] <= 20) & (df_ords_prods_merge_2['median_order_frequency'] > 10), 'order_frequency_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge_2.loc[df_ords_prods_merge_2['median_order_frequency'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```
